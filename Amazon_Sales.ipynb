```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/Amazon Sales.csv')
df=pd.DataFrame(df)
df
```

```python
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['ship-city'].fillna(df['ship-city'].mode()[0],inplace=True)
df['ship-state'].fillna(df['ship-state'].mode()[0],inplace=True)
df['ship-country'].fillna(df['ship-country'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Date']=pd.to_datetime(df['Date'])
df['Amount']=pd.to_numeric(df['Amount'],errors='coerce')
```

```python
df[['Date','Amount']]
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()

df.drop(columns=['index', 'Fulfilment', 'Sales Channel','ship-service-level','Style','ASIN','Courier Status','currency','ship-country',],inplace=True,errors='ignore')
df
```

```python
df.info()
```

```python
total_amount=df['amount'].sum()
print(total_amount)
total_orders=df.shape[0]
print(total_orders)
ave_order_value=df['amount'].mean()
print(ave_order_value)
```

```python
df = df[df['amount'] >= 0]
```

```python
import matplotlib.pyplot as plt
yearly_sales=df.groupby(df['date'].dt.to_period('Y'))['amount'].sum()
yearly_sales=yearly_sales.plot(kind='barh',figsize=(8,5),title='Yearly sales')
plt.grid(True)
plt.show(yearly_sales)

monthly_sales=df.groupby(df['date'].dt.to_period('M'))['amount'].sum().plot(kind='barh',figsize=(10,20),title='Monthly sales')
plt.grid(True)
plt.show(monthly_sales)

top_sales=df.groupby('ship-state')['amount'].sum().sort_values(ascending=False).head(10).plot(kind='pie',autopct='%1.1f%%',figsize=(6,6),title='Top sales in state')
plt.tight_layout()
plt.show(top_sales)

least_sales=df.groupby('ship-state')['amount'].sum().sort_values(ascending=False).tail(5).plot(kind='bar',figsize=(8,5),title='Top sales in state')
plt.grid(True)
plt.show(least_sales)
```

```python
# B2B
comparison = df.groupby('b2b')['amount'].agg(['mean', 'sum', 'count'])
print(comparison)

B2B=df['b2b'].value_counts().plot(kind='pie',autopct='%1.1f%%',figsize=(5,5),title='B2B')
plt.show(B2B)

fulifilled=df['fulfilment'].value_counts().plot(kind='bar',figsize=(3,3),title='% of orders fulfilled on time')
plt.show(fulifilled)
```

```python
category_wise=df.groupby('category')['amount'].sum().plot(kind='bar',figsize=(8,5),title='Category wise sales')
plt.grid(True)
plt.show(category_wise)
```

```python
status = df['status'].value_counts().plot(kind='bar',color='blue',figsize=(8,5),title='Status of Orders')
plt.xlabel('Order Status')
plt.ylabel('Count')
plt.show(status)
```

```python
import seaborn as sns
plt.figure(figsize=(10,5))

Order_Values=sns.boxplot(x=df['amount'],color="skyblue",fliersize=5,linewidth=2)

# Add mean line
plt.axvline(df['amount'].mean(), color='red', linestyle='--', label=f"Mean: {df['amount'].mean():.2f}")

# Titles & labels
plt.title('Distribution of Order Values (Box Plot)', fontsize=14, fontweight='bold')
plt.xlabel('Order Value (₹)', fontsize=12)
plt.legend()

plt.show(Order_Values)
```

```python
category_aov = df.groupby('category')['amount'].mean().sort_values(ascending=False).plot(kind='bar', figsize=(10,5), color='orange')
plt.title('Average Order Value by Category')
plt.xlabel('Category')
plt.ylabel('AOV (₹)')
plt.show(category_aov)
```

```python
size_sales=df.groupby('size')['amount'].sum().plot(kind='bar',figsize=(8,5),title='Size wise sales')
plt.grid(True)
plt.show(size_sales)

sales_by_city=df.groupby('ship-city')['amount'].sum().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(8,5),title='Sales by city')
plt.grid(True)
plt.show(sales_by_city)
```
